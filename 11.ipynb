```python
plt.figure(figsize=(4,3))
sns.heatmap(ANI_num_data.isnull())
```

NO null values in the Arunachal Pradesh's dataframe

```python
fd = pd.DataFrame(ANI_num_data)
to_convert = [
    "pH", "EC (µS/cm at)", "Ca (mg/L)", "Cl (mg/L)", "F (mg/L)", "Fe (ppm)", "HCO3", "K (mg/L)",
    "Mg (mg/L)", "NO3", "Na (mg/L)", "PO4", "SO4", "Total Hardness"
]
for col in to_convert:
    fd[col] = pd.to_numeric(fd[col], errors="coerce")
print(fd.dtypes)
```

```python
mean_values = [fd[col].mean() for col in to_convert]
mean_values
```

```python
skewness_values = {col: skew(fd[col].dropna()) for col in to_convert}
skewness_values
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
fd[to_convert] = scaler.fit_transform(fd[to_convert])
```

```python

plt.figure()
sns.displot(skewness_values)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.kdeplot(skewness_values, fill=True)
plt.show()
```

```python
from sklearn.preprocessing import PowerTransformer

pt = PowerTransformer(method="yeo-johnson")  # Auto-adjusts transformation
fd[to_convert] = pt.fit_transform(fd[[to_convert]])
```

if there is any duplicate entries

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)  # Reduce to 2 dimensions
X_pca = pca.fit_transform(fd)  # Transform dataset

# Plot first two principal components
plt.scatter(X_pca[:, 0], X_pca[:, 1])
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("PCA Reduced Data")
plt.show()
```
