**Interpretation of Colors:**
- ð´ Dark Red (Close to +1) â Strong Positive Correlation

Example: "Mg (mg/L)" and "Ca (mg/L)" (Correlation â 0.93)

This means that as one variable increases, the other also increases.

- ðµ Dark Blue (Close to -1) â Strong Negative Correlation

Example: "pH" and "Total Hardness" (Correlation â -0.17)

This means that as one variable increases, the other decreases.

- âª Light Colors (Near 0) â Weak or No Correlation

Example: "NO3" and "SO4" (Correlation â 0.06)

This suggests that these variables are not strongly related.

**1ï¸â£ Identify Highly Correlated Variables (ð´ > 0.8 or ðµ < -0.8)**
Highly positive correlations suggest redundancy (e.g., "Ca (mg/L)" and "Mg (mg/L)" with 0.93).

Possible Action: Consider removing one of the strongly correlated features to avoid multicollinearity (if using regression models).

**â Variables to consider for removal or combination:**

"Ca (mg/L)" and "Mg (mg/L)"

"EC (ÂµS/cm at)" and "Total Hardness" (â 0.83)

"Na (mg/L)" and "EC (ÂµS/cm at)" (â 0.74)

```python
# Load your dataset (Assuming 'df' is your cleaned data after normalization)
# df = pd.read_csv("your_data.csv")  # Uncomment if loading from a file

# Compute the correlation matrix
correlation_matrix = fd.corr().abs()  # Get absolute correlations

# Set threshold for high correlation (e.g., 0.8)
threshold = 0.8

# Find pairs of highly correlated variables
high_corr_pairs = np.where(correlation_matrix > threshold)
high_corr_pairs = [(correlation_matrix.index[i], correlation_matrix.columns[j]) 
                   for i, j in zip(*high_corr_pairs) if i != j]

# Convert list to a set to remove duplicates
unique_high_corr_vars = set([pair[1] for pair in high_corr_pairs])

# Drop one variable from each highly correlated pair
fd_reduced = fd.drop(columns=unique_high_corr_vars)

print(f"Removed variables due to high correlation: {unique_high_corr_vars}")
print(f"Shape before: {df.shape}, After: {fd_reduced.shape}")

# Save the cleaned dataset
# fd_reduced.to_csv("cleaned_data.csv", index=False)
fd_reduced
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)  # Reduce to 2 dimensions
X_pca = pca.fit_transform(fd_reduced)  # Transform dataset

# Plot first two principal components
plt.scatter(X_pca[:, 0], X_pca[:, 1])
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("PCA Reduced Data")
plt.show()
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
X_pca = pca.fit_transform(fd)  # without reducing according to the correlation

# Plot first two principal components
plt.scatter(X_pca[:, 0], X_pca[:, 1])
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("PCA Reduced Data")
plt.show()
```

```python
# Check explained variance ratio
explained_variance = pca.explained_variance_ratio_
print("Explained Variance Ratio:", explained_variance)
print("Total Variance Explained:", sum(explained_variance))
```


Since total variance is not above 85-90% (only 45% is retained). So we will n_components = 0.9 means keep 90% of variance.

```python

pca = PCA(n_components=0.9)  #keep 90% variance
X_pca = pca.fit_transform(fd_reduced)  # Transform dataset of reduced dataset

# Plot first two principal components
plt.scatter(X_pca[:, 0], X_pca[:, 1])
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("PCA Reduced Data")
plt.show()
```

```python
explained_variance = pca.explained_variance_ratio_
print("Explained Variance Ratio:", explained_variance)
print("Total Variance Explained:", sum(explained_variance))
```

Now, variance obtained is above 85-90% means PCA transformation retains most of the original dataset's information
